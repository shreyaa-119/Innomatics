```python
from langchain_core.prompts import PromptTemplate

# Define templates
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step."
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Generate 3 interview questions about {topic}."
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a short engaging story."
)

# Function to run all variations
def run_prompt_variations(topic):
    print("=== Prompt Variations Output ===\n")

    print("1. Teaching Style:")
    print(teaching_template.format(topic=topic))
    print()

    print("2. Interview Style:")
    print(interview_template.format(topic=topic))
    print()

    print("3. Storytelling Style:")
    print(story_template.format(topic=topic))
    print()

# Test
run_prompt_variations("Machine Learning")
```

```python
from langchain_core.prompts import ChatPromptTemplate

def get_chat_prompt(role, topic):

    role_map = {
        "teacher": "You are a helpful teacher who explains concepts clearly.",
        "interviewer": "You are an interviewer asking challenging questions.",
        "motivator": "You inspire and motivate learners."
    }

    system_message = role_map.get(role, "You are a helpful assistant.")

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Tell me about {topic}")
    ])

    messages = chat_template.format_messages(topic=topic)

    return messages

# Clean display function
def display_chat(messages):
    print("=== Chat Output ===\n")

    for msg in messages:
        role = msg.type.upper()
        content = msg.content
        print(f"{role}: {content}\n")

# Test
messages = get_chat_prompt("teacher", "Neural Networks")
display_chat(messages)
```

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Validation flags
    audience_valid = audience in valid_audience
    tone_valid = tone in valid_tone

    # Apply defaults if invalid
    if not audience_valid:
        print("Invalid audience detected → defaulting to 'beginner'")
        audience = "beginner"

    if not tone_valid:
        print("Invalid tone detected → defaulting to 'formal'")
        tone = "formal"

    return audience, tone


# Test cases
print("=== Validation Output ===\n")

a, t = validate_inputs("kids", "friendly")
print(f"Final Audience: {a}")
print(f"Final Tone: {t}")
```

```python
from langchain_core.prompts import PromptTemplate

def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate inputs
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Style-based templates
    style_map = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style.",
        "interview": "Generate interview questions about {topic} for {audience} in a {tone} tone.",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style."
    }

    selected_template = style_map.get(style, style_map["teaching"])

    # Step 3: Create PromptTemplate
    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template
    )

    # Step 4: Generate output
    output = template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return output


# Test
print("=== Prompt Generator Output ===\n")

result = generate_prompt("Neural Networks", "beginner", "fun", "storytelling")
print(result)
```

```python
from langchain_core.prompts import PromptTemplate

# Single reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone."
)

# Test inputs
test_cases = [
    ("Artificial Intelligence", "beginner", "fun"),
    ("Python Programming", "intermediate", "formal"),
    ("Data Science", "expert", "casual"),
    ("Machine Learning", "beginner", "formal"),
    ("Deep Learning", "expert", "fun")
]

print("=== Reusability Test Output ===\n")

# Run all cases
for i, (topic, audience, tone) in enumerate(test_cases, start=1):
    print(f"Test Case {i}:")
    output = reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )
    print(output)
    print()
```
