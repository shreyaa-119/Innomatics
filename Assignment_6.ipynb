```python
def smart_parking(capacity, logs):
    parked = 0
    peak = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        peak = max(peak, parked)

    print("Currently Parked Vehicles:", parked)

    if parked > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)



smart_parking(50, ["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
def estimate_delivery(distance, traffic, weather):

    time = distance * 5


    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8

    if weather == "Rainy":
        time += 10
    elif weather == "Storm":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")



estimate_delivery(8, "High", "Rainy")
```

```python
def seat_occupancy(total_seats, booked_list):
    booked = len(booked_list)
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Open Additional Show")
    else:
        print("Show Status: Seats Available")



seat_occupancy(200, [1]*150)
```

```python
def server_load(cpu_readings):
    total = 0

    for value in cpu_readings:
        total += value

    avg = total / len(cpu_readings)

    print("Average CPU Load:", int(avg), "%")

    if avg < 50:
        print("Server Status: Normal")
    elif avg <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")



server_load([45, 60, 70, 85, 90])
```

```python
def resource_monitor(resources):
    overused = []

    for resource, hours in resources.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")



resource_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
def event_registration(capacity, registrations):
    confirmed = min(capacity, registrations)
    waitlisted = max(0, registrations - capacity)

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if registrations >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")



event_registration(100, 105)
```
