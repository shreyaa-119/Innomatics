```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Recharge must be at least ₹50. Try again.")
            continue

        if amount not in valid_plans:
            print("Invalid plan! Available plans:", valid_plans)
            continue

        print("Recharge Successful ✅")
        break
```

```python
def inventory_alert(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert ⚠️")
        else:
            print(f"{product}: Stock OK")

products = {
    "Keyboard": 20,
    "Mouse": 10,
    "Monitor": 8,
    "USB Cable": 25
}

print("\nProblem 2 Output:")
inventory_alert(products)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak.lower() == "yes":
        fare *= 1.25   # 25% extra

    return fare

def cab_fare_system():
    while True:
        try:
            km = float(input("\nEnter distance (km): "))
            peak = input("Peak hour? (yes/no): ")

            total_fare = calculate_fare(km, peak)
            print(f"Total Fare: ₹{total_fare:.2f}")

            retry = input("Calculate again? (yes/no): ")
            if retry.lower() != "yes":
                break

        except ValueError:
            print("Invalid input! Try again.")
```

```python
def attendance_checker(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance = {percentage:.2f}% → Eligible"
    else:
        return f"Attendance = {percentage:.2f}% → Not Eligible"

print("\nProblem 5 Output:")
print(attendance_checker(["P","P","A","P","P","A","P","P"]))
```

```python
def check_password_strength(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password  (Minimum 8 characters required)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password  (Must include digit & special character)"

print("\nProblem 6 Output:")
print(check_password_strength("Pass@123"))
```
